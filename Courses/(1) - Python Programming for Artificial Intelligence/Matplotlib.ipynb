## Matplotlib

### LinePlot

```python
import matplotlib.pyplot as plt
```

```python
df1 = df.copy()
```

```python
df1 = df1.set_index("Id")
```

```python
df1.head()
```

### matplotlib.pyplot.plot

matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)[source]



```python
df1.plot()
```

```python
setosa = df1[df1.Species == "Iris-setosa"]
versicolor = df1[df1.Species == "Iris-versicolor"]
virginica = df1[df1.Species == "Iris-virginica"]
```

```python
plt.plot(setosa.index, setosa.PetalLengthCm, color = "red", label="setosa")
plt.plot(versicolor.index, versicolor.PetalLengthCm, color = "green", label="versicolor")
plt.plot(virginica.index, virginica.PetalLengthCm, color = "blue", label="virginica")
plt.xlabel("Id")
plt.ylabel("Petal Length")
plt.legend()
plt.show()
```

The legend module defines the Legend class, which is responsible for drawing legends associated with axes and/or figures.

```python
df1.plot(grid=True, linestyle = ":")
# grid ile gsterme
# linestyle : grafii gsterirken nasl gsterecek?
```

```python
df1.plot(grid=True, alpha = 0.5) #alpha: saydamlk
```

### Scatter Plot

It use to compare two different variables.

```python
plt.scatter(x=setosa.PetalLengthCm, y=setosa.PetalWidthCm, color = "red",label="setosa")
plt.scatter(x=versicolor.PetalLengthCm, y=versicolor.PetalWidthCm, color = "green",label="versicolor")
plt.scatter(x=virginica.PetalLengthCm, y=virginica.PetalWidthCm, color="blue",label="virginica")
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.title("ScatterPlot")
plt.legend()
plt.show()
```

### Histogram

```python
plt.hist(setosa.PetalLengthCm)
plt.xlabel("PetalLengthCm")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

### Bar Plot

```python
import numpy as np

x = np.array([1,2,3,7,4,9,6,3,5,7,4,2,8,3])
y = x*2 + 5
```

```python
plt.bar(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Plot")
plt.show()
```

### Subplots

```python
df1.plot(grid=True, alpha=0.9, subplots = True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(setosa.index,setosa.PetalLengthCm, color="red",label="setosa")
plt.ylabel("Setosa, PetalLengthCm")
plt.subplot(2,1,2)
plt.plot(versicolor.index,setosa.PetalLengthCm, color="blue",label="versicolor")
plt.ylabel("Versicolor, PetalLengthCm")
plt.legend()
plt.show()
```
